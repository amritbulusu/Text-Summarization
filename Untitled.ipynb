```python
def preprocess(text, stopwords_removal=False, stemming=False):
    cachedStopWords = stopwords.words("english")
    text = text.lower()
    text = re.sub(r'[_"\-;%()|+&=*%.,!?:#$@\[\]/]', ' ', text)
    text = re.sub(r" ' ", " ", text)
    str1 = re.findall("[@a-zA-Z0-9]* '[a-zA-Z]", text)
    for l in str1:
        text= re.sub(l,l.replace(" ",""),text)
    text = re.findall(r"[\w']+", text)
    new_text = []
    for word in text:
        if word in contractions:
            new_text.append(contractions[word])
        else:
            new_text.append(word)
    text = " ".join(new_text)
    tokens = text.split()
    words = []
    if stopwords_removal:
        for token in tokens:
            if token not in cachedStopWords:
                words.append(token)
            text = " ".join(words)        
    if stemming:
        text = [" ".join([stem(word) for word in sentence.split(" ")]) for sentence in text]
    return text
```

```python
for i in range(len(users["user_text"])):
    users["user_text"][i] = preprocess(users["user_text"][i], True)
```
